```python
merged_data.rename(columns={'AIRLINE_x': 'AIRLINE_CODE','AIRLINE_y': 'AIRLINE'}, inplace=True)
```

```python
df = pd.merge(merged_data, data_airports, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE', how='left')
```
